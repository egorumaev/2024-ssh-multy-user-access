Настроим подключение из ВМ client к ВМ server с использованием SSH-ключей.
Ключи могут быть сгенерированы с помощью различных алгоритмов, которые поддерживает текущая версия протокола SSH. Например, если использован тип шифрования RSA, то файлы будут именоваться следующим образом [9]:

– id_rsa – закрытый ключ;

– id_rsa.pub – публичный (открытый) ключ.


* [9] Авторизация на базе ключей SSH // Selectel (Серверы. Облако. Дата-центры). –  Режим доступа: https://selectel.ru/blog/ssh-authentication/ (дата обращения 18.10.2024)

Открытый SSH-ключ хранится на удаленном узле, его можно размещать в публичном доступе, т.к. наличие одного лишь публичного ключа не позволяет злоумышленнику расшифровать перехваченную информацию. Закрытый (приватный) SSH-ключ является ключом к данным. С его помощью могут быть расшифрованы сами сообщения. Приватный ключ ни в коем случае нельзя передавать в открытый (публичный) доступ. Он должен храниться только у владельца. Для дополнительной защиты на закрытый ключ может быть установлен пароль. Проверим, какие алгоритмы шифрования доступны.

````python
```bash
ssh-keygen ?
```
````

Picture SSH-keygen.png

На ВМ client создадим ключи для подключения. Для шифрования доступны шесть алгоритмов: dsa; ecdsa; ecdsa-sk; ed25519; ed25519-sk; rsa.

Алгоритм шифрования **RSA** имеет наилучшую совместимость, но для обеспечения повышенного уровня безопасности требуется большой размер ключа (варьируется от 1024 до 16384 бит), что повышает нагрузку на вычислительные мощности.

**DSA** в сравнении с RSA показывает лучшие показатели при генерации подписи, но уступает по времени при ее проверке. Максимальная длина ключа – 1024 бит, что делает алгоритм не самым безопасным от взлома.

Алгоритм **ECDSA** основан на использовании эллиптических кривых и модульной арифметике. Он превосходит алгоритм RSA в скорости, т.к. требуется ключ гораздо большего размера. Однако, он более уязвим перед взломом с помощью квантовых вычислений.

Алгоритм **ED25519** базируется на эллиптической кривой, которая дает уровень безопасности выше, чем алгоритмы ECDSA и DSA, и одновременно хорошую производительность. Недостатком алгоритма является то, что он поддерживается в версиях от OpenSSH 6.5 и выше.

**ECDSA-SK** и **ED25519-SK** – это улучшенные версии алгоритмов ECDSA и ED25519, которые доступны в версии OpenSSH 8.2 и выше.

* **(5)** Дополнительно на ВМ **server** настроен способ доступа к SSH-серверу по ключам, сгенерированным с помощью протокола шифрования **ecdsa**, что позволяет организовывать защищенное от действий злоумышленников подключение от ВМ **server** к другим ВМ созданной локальной сети.
